```python
# Guarda todas las IDs de pedidos que tengan un valor ausente en 'add_to_cart_order'
pedidos_con_ausentes = order_products[order_products['add_to_cart_order'].isnull()]['order_id'].unique()
print(f"IDs de pedidos con valores ausentes: {len(pedidos_con_ausentes)} pedidos únicos")
```

```python
# ¿Todos los pedidos con valores ausentes tienen más de 64 productos?
pedidos_ausentes_agrupados = order_products[order_products['add_to_cart_order'].isnull()].groupby('order_id')

# Agrupa todos los pedidos con datos ausentes por su ID de pedido.
conteo_productos = pedidos_ausentes_agrupados['product_id'].count()

# Cuenta el número de 'product_id' en cada pedido y revisa el valor mínimo del conteo.
print(f"Número mínimo de productos en pedidos con valores ausentes: {conteo_productos.min()}")
print(f"Número máximo de productos en pedidos con valores ausentes: {conteo_productos.max()}")
print(f"¿Todos los pedidos tienen más de 64 productos? {(conteo_productos > 64).all()}")
```

```python
# Código del revisor
order_products[order_products['order_id'].isin(pedidos_con_ausentes)].groupby('order_id')['product_id'].count().min()
```

Describe brevemente cuáles son tus hallazgos.
- 836 valores ausentes (0.02% del total)
- Afectan a 70 pedidos únicos
- Rango válido: 1-64 productos por carrito
¿Los valores ausentes se deben a pedidos con más de 64 productos? NO
- Pedidos con valores ausentes: 1-63 productos (máximo) > Ninguno supera los 64 productos
Los valores ausentes no se deben a limitaciones técnicas del sistema, sino probablemente a errores en la captura de datos durante el proceso de compra.

```python
# Remplaza los valores ausentes en la columna 'add_to_cart? con 999 y convierte la columna al tipo entero.
order_products['add_to_cart_order'] = order_products['add_to_cart_order'].fillna(999).astype(int)
# Para verificar el resultado
print("Valores únicos después del reemplazo:")
print(order_products['add_to_cart_order'].value_counts().sort_index().tail(10))
print(f"\nTipo de datos: {order_products['add_to_cart_order'].dtype}")
print(f"Valores ausentes restantes: {order_products['add_to_cart_order'].isnull().sum()}")
```

Describe brevemente tus hallazgos y lo que hiciste con ellos.

### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
# Calcular estadísticas por producto
estadisticas_producto = order_products.groupby('product_id').agg({
    'reordered': ['count', 'sum']  # count = total pedidos, sum = total reordenados
}).reset_index()

# Aplanar nombres de columnas
estadisticas_producto.columns = ['product_id', 'total_pedidos', 'total_reordenados']

# Calcular tasa de reorden
estadisticas_producto['tasa_reorden'] = (
    estadisticas_producto['total_reordenados'] / estadisticas_producto['total_pedidos']
)

print("Primeras 10 filas de estadísticas por producto:")
print(estadisticas_producto.head(10))
```

```python
# Productos con mayor y menor tasa de reorden
print("Productos con mayor tasa de reorden:")
top_tasas = estadisticas_producto.nlargest(10, 'tasa_reorden')
print(top_tasas[['product_id', 'total_pedidos', 'total_reordenados', 'tasa_reorden']])

print("\nProductos con menor tasa de reorden:")
bottom_tasas = estadisticas_producto.nsmallest(10, 'tasa_reorden')
print(bottom_tasas[['product_id', 'total_pedidos', 'total_reordenados', 'tasa_reorden']])
```

```python
# Crear histograma de distribución de tasas de reorden
plt.figure(figsize=(12, 6))
plt.hist(estadisticas_producto['tasa_reorden'], bins=50, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tasas de Reorden por Producto')
plt.xlabel('Tasa de Reorden (proporción)')
plt.ylabel('Número de Productos')
plt.grid(axis='y', alpha=0.3)
plt.show()

# Estadísticas descriptivas
print("Estadísticas de tasas de reorden:")
print(estadisticas_producto['tasa_reorden'].describe())
```

Escribe aquí tus conclusiones

### [C4] Para cada cliente, ¿qué proporción de sus productos ya los había pedido?

```python
# Unir order_products con orders para obtener user_id
order_products_con_usuarios = order_products.merge(
    orders_clean[['order_id', 'user_id']],
    on='order_id',
    how='left'
)

# Calcular estadísticas por cliente
estadisticas_cliente = order_products_con_usuarios.groupby('user_id').agg({
    'reordered': ['count', 'sum']  # count = total productos, sum = total reordenados
}).reset_index()

# Aplanar nombres de columnas
estadisticas_cliente.columns = ['user_id', 'total_productos', 'productos_reordenados']

print("Primeras 10 filas de estadísticas por cliente:")
print(estadisticas_cliente.head(10))
print(f"\nTotal de clientes analizados: {len(estadisticas_cliente):,}")
```

```python
# Calcular proporción de productos reordenados por cliente
estadisticas_cliente['proporcion_reorden'] = (
    estadisticas_cliente['productos_reordenados'] /
    estadisticas_cliente['total_productos']
)

print("Estadísticas de proporción de reorden por cliente:")
print(estadisticas_cliente['proporcion_reorden'].describe())

print(f"\nCliente con mayor proporción de reorden: {estadisticas_cliente['proporcion_reorden'].max():.4f}")
print(f"Cliente con menor proporción de reorden: {estadisticas_cliente['proporcion_reorden'].min():.4f}")

# Mostrar algunos ejemplos
print("\nPrimeros 10 clientes y sus proporciones:")
print(estadisticas_cliente[['user_id', 'total_productos', 'productos_reordenados', 'proporcion_reorden']].head(10))
```
